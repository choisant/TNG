```python
import h5py
from sphviewer.tools import QuickView
import matplotlib.pyplot as plt

with h5py.File('dm_halo.h5py', 'r') as f:
    pos = f['Coordinates'].value

qv = QuickView(pos, r='infinity', plot=False)

hsml = qv.get_hsml()

qv.imshow(cmap="gist_stern", vmin=2)
img = qv.get_image()
extent = qv.get_extent()
fig = plt.figure(1, figsize=(10,10))
plt.imshow(img, extent=extent, cmap="gist_heat")
```

```python
qv_subhalo = QuickView(pos, hmsl=hsml, r='infinity', plot=False, 
                       x=-0.005, y=-0.045, z=0, extent=[-0.02, 0.02, -0.02, 0.02])
fig = plt.figure(1, figsize=(10, 10))
plt.imshow(qv_subhalo.get_image(), extent=qv_subhalo.get_extent(),
      cmap='gist_heat', origin='lower')
plt.show()
```

```python
qv = QuickView(pos, r='0.1', plot=False,
              x=0, y=0, z=0, extent=[-0.1, 0.1, -0.1, 0.1])

hsml = qv.get_hsml()

qv.imshow(cmap="gist_stern", vmin=2)
img = qv.get_image()
extent = qv.get_extent()
fig = plt.figure(1, figsize=(10,10))
plt.imshow(img, extent=extent, cmap="gist_heat", origin='lower')
```
